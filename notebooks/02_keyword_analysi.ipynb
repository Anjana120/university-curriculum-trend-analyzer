```python
import pandas as pd

df = pd.read_csv("../data/mit_courses.csv")
df.head()
```

```python
keywords = ["python", "machine learning", "data", "ai", "statistics", "cloud", "blockchain", "sql", "nlp"]
```

```python
from collections import Counter

# Combine all course descriptions into one large string
all_text = " ".join(df["description"].dropna()).lower()

# Count how often each keyword appears
keyword_counts = {kw: all_text.count(kw.lower()) for kw in keywords}

# Show sorted result
sorted(keyword_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
import matplotlib.pyplot as plt

# Convert dict to sorted lists
words = list(keyword_counts.keys())
counts = list(keyword_counts.values())

plt.figure(figsize=(10,5))
plt.bar(words, counts, color="skyblue")
plt.title("Keyword Mentions in MIT Course Descriptions")
plt.xlabel("Keyword")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
